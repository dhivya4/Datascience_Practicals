```python
data.drop_duplicates(inplace = True)
```

```python
data['gender'].unique()
```

```python
# filling series with max values
data['gender'] =  data['gender'].fillna(data['gender'].value_counts().idxmax())
```

```python
# filling series with max values
data['enrolled_university'] =  data['enrolled_university'].fillna(data['enrolled_university'].value_counts().idxmax())
```

```python
data['experience'].value_counts()
```
